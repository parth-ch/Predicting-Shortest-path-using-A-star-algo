```python
def get_data_for_buses(start,end):
    url = f'https://www.makemytrip.com/bus/search/{start}/{end}/06-11-2024'
    print(url)
    driver = webdriver.Chrome()  
    driver.get(url)
    time.sleep(2)
    source = driver.page_source
    driver.quit()
    soup = BeautifulSoup(source,'html.parser')
    buses = soup.find_all('div',{'class':'busCardContainer'})
        
    busdf = {
        'travel_name' : [],
        'start_times':[],
        'travel_times':[],
        'end_times':[],
        'price':[],
        'type_route':[],
        'available_seats':[],
        'window_seats':[],
        'rating':[]
    }

    for b in buses:
        name = b.find('p',{'class':'makeFlex hrtlCenter appendBottom8 latoBold blackText appendRight15'}).text
        start_times = b.find('span',{'class':'font18 latoBlack blackText appendRight4'}).text
        travel_times = b.find('div',{'class':'font14 secondaryTxt'}).text
        end_times = b.find('span',{'class':'font18 blackText appendRight4 latoRegular'}).text
        price = b.find('span',{'id':'price'}).text
        type_route = b.find('p',{'class':'makeFlex hrtlCenter secondaryTxt nowrapStyle'}).text
        seats_av = b.find_all('ul',{'class':'sc-fjdhpX fXgCif'})
        available_seats = [ seats_av[i].text for i in range(len(seats_av)) if i % 2 == 0 ]
        window_seats = [ seats_av[i].text for i in range(len(seats_av)) if i % 2 == 1 ]
        rating = b.find('span',{'class':'sc-kGXeez bfoMHl'})
        if rating is not None:
            rating = rating.text
        else:
            rating = None
        busdf['travel_name'].append(name)
        busdf['start_times'].append(start_times)
        busdf['travel_times'].append(travel_times)
        busdf['end_times'].append(end_times)
        busdf['price'].append(price)
        busdf['type_route'].append(type_route)
        busdf['available_seats'].append(available_seats)
        busdf['window_seats'].append(window_seats)
        busdf['rating'].append(rating)
        
    df = pd.DataFrame(busdf)
    return df
```

```python
get_data_for_buses('nashik','mumbai')
```

```python
cities = pd.read_csv('indian-cities-dataset.csv')
cities
```

```python
megadf = pd.DataFrame()
i = 0
for start,end,distance in zip(cities['Origin'],cities['Destination'],cities['Distance']):
    df = get_data_for_buses(start,end)
    df['start'] = start
    df['end'] = end
    df['distance'] = distance
    megadf = pd.concat([megadf,df])
    print(i)
    i = i+1
```

```python
megadf.to_csv('buses.csv')
```
